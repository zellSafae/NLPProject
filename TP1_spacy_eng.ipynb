```python
# *****************zellou safae - TP1 -  Processing texts using spaCy library ****************************
!pip install spacy
# Import the spaCy library
import spacy
# Load the small language model for English and assign it to the variable 'nlp'
nlp = spacy.load('en_core_web_sm')
# Call the variable to examine the object :  returns a spaCy Language object that contains a language model for the English language
nlp
#assign the english test to a variable
Engtext="""Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely
the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from
the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational
numbers, irrational numbers, geometrical magnitudes, etc., to all be treated as "algebraic objects". It gave mathematics a whole new
development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the
subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a
way which had not happened before."""
#calling the variable to examine the output 
Engtext
#feed the string object under 'Engtext' to the language object under 'nlp'
#store  the result under doc 
doc=nlp(Engtext)
#call the variable to examine the object 
doc
print("********************")
print(doc)
```

```python
#Tokenization :  breaking the text down into analytical units in need of further processing.
#The output shows one Token per line
#Punctuation marks such as ‘.’ and ‘,’ constitute their own Tokens
for token in doc:
    print(token)
```

```python
#Part-of-speech tagging :  determining the word class of a token
#To access the results of POS tagging, let’s loop over the Doc object doc and print each Token and its coarse and fine-grained part-of-speech tags
for token in doc:
    print(token,token.pos_,token.tag_)
```

```python
#Syntactic parsing
for token in doc:
    print(token,token.dep_)
#print the index of current token ,the token,the token itself,the dependency,the dependency, the head and its index
    print(token.i,token,token.dep_,token.head.i,token.head)
```

```python
#Sentence segmentation
# Loop over sentences in the Doc object and count them using enumerate()
for number, sent in enumerate(doc.sents):
    
    # Print the token and its dependency tag
    print(number, sent)
```

```python
#Lemmatization
#A lemma is the base form of a word,Lemmas are available for each Token under the attribute lemma_
# Loop over items in the Doc object, using the variable 'token' to refer to items in the list
for token in doc:
    
    # Print the token and its dependency tag
    print(token, token.lemma_)
```
